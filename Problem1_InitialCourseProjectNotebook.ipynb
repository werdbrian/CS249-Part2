```python
%load_ext rmagic
import rpy2 as Rpy
import json
import pandas as pd
from glob import glob
```

```python
'''
Convert Yelp Academic Dataset from JSON to CSV
 
Requires Pandas (https://pypi.python.org/pypi/pandas)
 
By Paul Butler, No Rights Reserved
'''
 
def convert(x):
    ''' Convert a json string to a flat python dictionary
    which can be passed into Pandas. '''
    ob = json.loads(x)
    for k, v in ob.items():
        if isinstance(v, list):
            ob[k] = ','.join(str(v))
        elif isinstance(v, dict):
            for kk, vv in v.items():
                ob['%s_%s' % (k, kk)] = vv
            del ob[k]
    return ob
 
for json_filename in glob('*.json'):
    csv_filename = '%s.csv' % json_filename[:-5]
    print 'Converting %s to %s' % (json_filename, csv_filename)
    df = pd.DataFrame([convert(line) for line in file(json_filename)])
    df.to_csv(csv_filename, encoding='utf-8', index=False)
```

#Reviews Dataset Analysis

```python
%%R

# Load the reviews data (335,022)
data_review = read.csv("yelp_academic_dataset_review.csv", stringsAsFactors=FALSE)
print(dim(data_review))
```

```python
%%R -w 800

library(MASS)

data_review$length <- (nchar((data_review$text)))
# Efficient way to replace non-positive values with 1
data_review$length <- (data_review$length + abs(data_review$length) + 2) / 2
cat("New matrix dimension (after adding length column): "); print(dim(data_review))

# Separate review dataset based on star rating
data_review_1stars <- subset(data_review, stars==1)
data_review_2stars <- subset(data_review, stars==2) 
data_review_3stars <- subset(data_review, stars==3)
data_review_4stars <- subset(data_review, stars==4)
data_review_5stars <- subset(data_review, stars==5)

# Look at rating (stars) distribution
# Distribution is obviously skewed. People tend to write positive reviews
counts <- c(dim(data_review_1stars)[1], dim(data_review_2stars)[1], dim(data_review_3stars)[1], 
            dim(data_review_4stars)[1], dim(data_review_5stars)[1])
barplot(counts, xlab="Star Rating", ylab="Count", main="Number of Star Ratings", col="blue")

# Correlation between review length and star rating?
hist(data_review_1stars$length, xlab="Length", ylab="Count", main="Count of 1-Stars Review Lengths", 
     breaks=50, col="blue", prob=TRUE)
param_1star <- fitdistr(data_review_1stars$length, "log-normal")
curve(dlnorm(x, meanlog=param_1star$estimate[1], sdlog=param_1star$estimate[2]), col="red", add=TRUE)
legend("right", c("Log-normal"), col=c("red"), lwd=3)

hist(data_review_3stars$length, xlab="Length", ylab="Count", main="Count of 3-Stars Review Lengths", 
     breaks=50, col="blue", prob=TRUE)
param_3star <- fitdistr(data_review_3stars$length, "log-normal")
curve(dlnorm(x, meanlog=param_3star$estimate[1], sdlog=param_3star$estimate[2]), col="red", add=TRUE)
legend("right", c("Log-normal"), col=c("red"), lwd=3)

hist(data_review_5stars$length, xlab="Length", ylab="Count", main="Count of 5-Stars Review Lengths", 
     breaks=50, col="blue", prob=TRUE)
param_5star <- fitdistr(data_review_5stars$length, "log-normal")
curve(dlnorm(x, meanlog=param_5star$estimate[1], sdlog=param_5star$estimate[2]), col="red", add=TRUE)
legend("right", c("Log-normal"), col=c("red"), lwd=3)
```

We expected that the distribution of review lengths would differ more between the different star ratings. Inituitively, we thought 1-star and 5-star reviews would be longer because the reviewers feel strongly about one side and would write longer reviews to express their opinion. However, the data shows otherwise, and the distributions are similar distribution between the different star ratings.
