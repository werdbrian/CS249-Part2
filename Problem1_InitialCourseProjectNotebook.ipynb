```python
'''
Convert Yelp Academic Dataset from JSON to CSV
 
Requires Pandas (https://pypi.python.org/pypi/pandas)
 
By Paul Butler, No Rights Reserved
'''
 
def convert(x):
    ''' Convert a json string to a flat python dictionary
    which can be passed into Pandas. '''
    ob = json.loads(x)
    for k, v in ob.items():
        if isinstance(v, list):
            ob[k] = ','.join(str(v))
        elif isinstance(v, dict):
            for kk, vv in v.items():
                ob['%s_%s' % (k, kk)] = vv
            del ob[k]
    return ob
 
for json_filename in glob('*.json'):
    csv_filename = '%s.csv' % json_filename[:-5]
    print 'Converting %s to %s' % (json_filename, csv_filename)
    df = pd.DataFrame([convert(line) for line in file(json_filename)])
    df.to_csv(csv_filename, encoding='utf-8', index=False)
```

##Topic Modeling 
What do people tend to write about in their reviews? We will use each review as a document to form a corpus to perform LDA (Latent Dirichlet allocation) for topic modeling.

```python
%%R

library(tm)
library(topicmodels)
```

```python
%%R

data_review = read.csv("yelp_academic_dataset_review.csv", stringsAsFactors=FALSE)

data_review_subset <- subset(data_review, (stars == 1 | stars == 5) & nchar(toString(text)) > 100, c(text))

# Throw away reviews that are less than specified number of characters
data_review_5stars <- subset(data_review, stars == 5 & nchar(toString(text)) > 100, c(text))
data_review_1stars <- subset(data_review, stars == 1 & nchar(toString(text)) > 100, c(text))
cat("Number of 1 and 5 stars review: "); print(dim(data_review_subset))
cat("Number of 5 stars review: "); print(dim(data_review_5stars))
cat("Number of 1 stars review: "); print(dim(data_review_1stars))

# CHANGE sample size later!!
data_review_5stars_sample <- sample(data_review_5stars$text, 5000)
data_review_1stars_sample <- sample(data_review_1stars$text, 5000)
```

###Use Python to remove stopwords because 'tm' libray is limited.

```python
import re
import os
import pandas.rpy.common as com

#data_review_5stars_sample = com.load_data('data_review_5stars_sample')
data_review_subset = %Rget -d data_review_subset$text
data_review_5stars_sample = %Rget -d data_review_5stars_sample
data_review_1stars_sample = %Rget -d data_review_1stars_sample
print len(data_review_5stars_sample)

stoplist = set('also can don didn just s that the they this ive ve wouldn')

data_review_subset_processed = []
for doc in data_review_subset:
    # Remove punctuations
    doc = re.sub(r'[^a-zA-Z]', ' ', doc)
    # Remove stop words
    doc = [word for word in doc.lower().split() if word not in stoplist]
    data_review_subset_processed.append(' '.join(doc))

data_review_5stars_processed = []
for doc in data_review_5stars_sample:
    # Remove punctuations
    doc = re.sub(r'[^a-zA-Z]', ' ', doc)
    # Remove stop words
    doc = [word for word in doc.lower().split() if word not in stoplist]
    data_review_5stars_processed.append(' '.join(doc))

data_review_1stars_processed = []
for doc in data_review_1stars_sample:
    # Remove punctuations
    doc = re.sub(r'[^a-zA-Z]', ' ', doc)
    # Remove stop words
    doc = [word for word in doc.lower().split() if word not in stoplist]
    data_review_1stars_processed.append(' '.join(doc))

#%R data_review_5stars_sample <- sample(data_review_5stars$text, 100)
#%R print(data_review_5stars_sample)
#data_review_5stars_sample = %Rget -d data_review_5stars_sample

%R -i data_review_subset_processed
%R -i data_review_5stars_processed
%R -i data_review_1stars_processed
```

```python
%%R -w 800

corpus_subset <- Corpus(VectorSource(sample(data_review_subset_processed, 5000)))

# Remove stop words from text
more_stopwords <- c("the", "they", "that", "this", "also", "can", "don", "didn", "just", 
                    "will", "i've", "get", "one", "never", "really", "i'm")
corpus_subset <- tm_map(corpus_subset, removeWords, c(stopwords("english"), more_stopwords))

# Create doucment-term matrix
dtm_subset <- DocumentTermMatrix(corpus_subset)
cat("Size of document-term matrix: "); print(dim(dtm_subset))

# Train the model
k <- 20
lda_result_subset <- LDA(dtm_subset, k, method="VEM")

# Get beta, the logarithm of word distribution in topics
beta_subset <- attr(lda_result_subset,"beta")

# Get gamma, topic distribution for each document
gamma_subset <- attr(lda_result_subset, "gamma")
terms_subset <- terms(lda_result_subset, 15)

cat("Topics and terms for 1-star and 5-star reviews:\n"); print(terms_subset)
```
