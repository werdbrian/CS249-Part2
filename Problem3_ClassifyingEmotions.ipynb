# Problem 3:  Classifying Emotions

Classification is usually concerned with only a few classes, and many problems can arise when there are a number of classes.
Dealing with many classes is a research area in its own right.

The Mulan project at <a target="_blank" title="http://mulan.sourceforge.net" href="http://mulan.sourceforge.net">http://mulan.sourceforge.net</a> is one effort that studies classification with many classes.
Datasets for the project are available at <a target="_blank" title="http://mulan.sourceforge.net/datasets.html" href="http://mulan.sourceforge.net/datasets.html">http://mulan.sourceforge.net/datasets.html</a>.

Among these, the <em>emotions</em> dataset aims at classification of emotions present in music.
It is available at http://sourceforge.net/projects/mulan/files/datasets/emotions.rar
(For simplicity, it is also attached here on CourseWeb.)

The dataset is a 593x78 table describing 593 songs.
There are 72 numerical features for each piece describing their timbre and pitch,
followed by 6 different binary classifications:
<blockquote>
amazed_suprised, happy_pleased, relaxing_calm, quiet_still, sad_lonely, angry_aggressive
</blockquote>
One thing that makes the dataset particularly interesting
is that <i>there are 6 independent binary class variables</i>,
and so there are 2**6 = 64 possible classes.

Another way of putting this is that this is a <b>multi-label dataset</b>, with 6 independent emotion labels for each song.


```python
%%R

## Get an initial feel for the emotions dataset

emotions = read.csv("emotions.csv", header=TRUE)
print(dim(emotions))
p = ncol(emotions)

# increase line width for printing
options(width=125)

# extract only the six 0-1 classification columns at the end
classes = emotions[,(p-5):p]
cat(paste(c("\n\n",colnames(classes),"\n\n"),collapse="  "))

cat("\n\nThe 78 column names of the emotions table:\n\n")

print(colnames(emotions))
```

##The co-occurrence Patterns of Emotions in the Music Data have lots of structure

A little analysis confirms that the classifications data are significantly correlated,
and the emotion labels are anything but independent or uniformly distributed.

In fact there are only 27 distinct patterns (of 0s and 1s), and only 14 of these hae more than a few instances (out of 593).

Specifically, notice that the Table 1 that appears in paper [1] above is the same as the first table below:


```python
%%R -h 500 -w 500

# tabulate all patterns of 0's and 1's that actually occur
# by collapsing each row of classification values into a binary string

numbers_of_occurrences = apply(classes, 2, sum)
cat("\n\nnumbers of occurrences of each emotion label:\n\n")
print(t(t(numbers_of_occurrences)))

classification_patterns = apply(classes, 1, function(row) paste(row,collapse=""))
cat("\n\nnumbers of occurrences of each pattern of emotion labels:\n\n")
print(table(classification_patterns))

sorted_patterns = sort(table(classification_patterns))
sorted_labels = unlist(dimnames(sorted_patterns)[1])
np = length(sorted_patterns)
    
cat("\n\ntotal number of patterns:\n")
print(np)
                       
cat("\n\npatterns in order of frequency of occurrence:\n\n")
print(sorted_patterns)

plot(1:np, rev(sorted_patterns), pch=20, col="blue",
     main="numbers of occurrences of each emotion pattern", xlab="rank", ylab="number of occurrences")
text(1:np, rev(sorted_patterns), labels=rev(sorted_labels), pos=4, offset=0.5, col="blue", cex=0.65)

#  determine how the various classification values are correlated
heatmap( cor(classes), symm=TRUE, scale="none", margins=c(12,12))
   
cat("\n\ncorrelation matrix:\n\n")
print(round(cor(classes),2))
    
#  print the correlation table, permuted to match the heatmap
cat("\n\ncorrelation matrix with rows and columns in clustered order:\n\n")
print(round(cor(classes[,c(6,1:3,5,4)]),2))
```

##Developing Classifiers: Models of Musical Mood

<blockquote>
[2] G. Tsoumakas, I. Katakis, I. Vlahavas. <i>Random k-Labelsets for Multi-Label Classification</i>,
IEEE Transactions on Knowledge and Data Engineering, 23(7), 1079-1089, 2011.
http://mlkd.csd.auth.gr/publication_details.asp?publicationID=312
</blockquote>
This is an interesting paper also, reporting on many attempts to develop multi-label classifiers,
and proposing to use random subsets of labels.
