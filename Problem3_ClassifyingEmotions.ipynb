# Problem 3:  Classifying Emotions

Classification is usually concerned with only a few classes, and many problems can arise when there are a number of classes.
Dealing with many classes is a research area in its own right.

The Mulan project at <a target="_blank" title="http://mulan.sourceforge.net" href="http://mulan.sourceforge.net">http://mulan.sourceforge.net</a> is one effort that studies classification with many classes.
Datasets for the project are available at <a target="_blank" title="http://mulan.sourceforge.net/datasets.html" href="http://mulan.sourceforge.net/datasets.html">http://mulan.sourceforge.net/datasets.html</a>.

Among these, the <em>emotions</em> dataset aims at classification of emotions present in music.
It is available at http://sourceforge.net/projects/mulan/files/datasets/emotions.rar
(For simplicity, it is also attached here on CourseWeb.)

The dataset is a 593x78 table describing 593 songs.
There are 72 numerical features for each piece describing their timbre and pitch,
followed by 6 different binary classifications:
<blockquote>
amazed_suprised, happy_pleased, relaxing_calm, quiet_still, sad_lonely, angry_aggressive
</blockquote>
One thing that makes the dataset particularly interesting
is that <i>there are 6 independent binary class variables</i>,
and so there are 2**6 = 64 possible classes.

Another way of putting this is that this is a <b>multi-label dataset</b>, with 6 independent emotion labels for each song.


```python
%%R

## Get an initial feel for the emotions dataset

emotions = read.csv("emotions.csv", header=TRUE)
print(dim(emotions))
p = ncol(emotions)

# increase line width for printing
options(width=125)

# extract only the six 0-1 classification columns at the end
classes = emotions[,(p-5):p]
emotions_minus_classes = emotions[,1:(p-6)]
cat(paste(c("\n\n",colnames(classes),"\n\n"),collapse="  "))

cat("\n\nThe 78 column names of the emotions table:\n\n")

print(colnames(emotions))
print(dim(emotions_minus_classes))
```

##The co-occurrence Patterns of Emotions in the Music Data have lots of structure

In fact there are only 27 distinct patterns (of 0s and 1s), and only 14 of these have more than a few instances (out of 593).

Specifically, notice that the Table 1 that appears in paper [1] above is the same as the first table below:

## Methodology used to choose, train, and evaluate our model


To pick which model might model emotions the best, we first tried to model "angry_aggressive" 
using several methods, and after chosing a model (log-reg,lda,random forests), attempt to classify other emotions using this 
model, hoping to achieve comparable results to the average song-classifier. 

First we attempted to treat each song as a sextet of bits, and attempt to run linear regression on these sextets, treating each
sextet (e.g. 100010) as a factor.  With a reduced model, keeping only significant coefficients at the 10% level, we managed to 
build a model with an R^2 value of .2256 which suggested that this approach was not appropriate in modeling song emotions.

Given that we could not build one model which encompassed all the classes by concatenating the classes, we decided to build 
six individual classifiers, each classifier responsible for one emotion, e.g. "happy_pleased".  We could then aggregate 
the results from our six classifiers to give a full classification of the emotion of a new song.

First we tried a plain full linear model regressing on the feature "angry_aggressive" with the lm() function.
The initial summary which showed a poor fit through a very low R^2 value of .4662 which made us reject this method.

Next we tried logistic regression using the glm() function.  On our first attempt we trained using every feature, 
and achieved a mean testing accuracy of 57% after running 10-fold cross valuation 10 times for evaluation.  
After using the stepAIC() function to remove features from our model to reduce overfitting, and hand picking 
only coefficients that were statistically significant at the 10% level, we managed to increase the accuracy of prediction to 65%. 
Achieving such low accuracy with the best model we could build, we decided to try LDA next.  

We built our LDA model treating 0's or 1's in the angry_aggressive column as different classes , indicating 
whether a song was angry_aggressive or not.  Training on all the song features in the training set using lda(), 
and the predict() function to assess our model, we managed to achieved an  accuracy of 82% for predicting 
angry_aggressive emotions in a song  in our cross validation testing set.

Getting promising results for the angry_aggressive class,  we decided to use LDA to build all our models for classifying songs.  The code used to choose, train, build, and evaluate our models are in the following cells:

```python
%%R
cv_10_fold = function(lda.fit,entire_aa)
## This function performs 10 x 10-fold cross validation given an LDA fit and a dataset from which to subset from.
## It then prints the mean cross validation error calculated across the 10 iterations.
{
nrep = 10 # number of times we will run cross-validation on entire data set
cv.error = rep(0,nrep) #vector used to store avg of accuracy on each CV iteration
nfold = 10 # number of folds in our CV
for( irep in 1:nrep ){ #runs our CV 10 times
    ifold = sample( rep(1:10, length = nrow(entire_aa) ) )
    for( ifoldi in 1:nfold ){ #partitions our training set into 10 parts and then calculates accuracy for each one
        testset = (ifoldi==ifold)
        testing_aa_fold = entire_aa[testset,]
        predictions<-predict(lda.fit, testing_aa_fold)
        incorrect.prediction  <-  (predictions$class != testing_aa_fold[,73]) #73rd column is feature we are trying to predict
        m <- dim(testing_aa_fold)[1] #total songs we are trying to classify in this particular fold
        accuracy <- (m - sum(incorrect.prediction))/m
#        print(accuracy)
    }
    cv.error[irep] = mean( accuracy ) #mean across 10-fold CV
}
#print(cv.error)
print(mean(cv.error)) #mean across 10 time 10-fold CV
}
```

```python
%%R
make_lda_model = function(training_aa, feature_name,entire_aa,testing_aa)
## This function constructs and evaluates an LDA model, given a class name, training, testing, and entire dataset.
## This function calls the cv_10_fold function defined earlier to perform cross-validation 10 times using 10 folds
{
    lda.fit <- lda(training_aa[,73]~., training_aa[,1:72]) # train last column using first 72 columns
    predictions<-predict(lda.fit, testing_aa)
    incorrect.prediction  <-  (predictions$class != testing_aa[,73])
    #print(feature_name)
    m <- dim(testing_aa)[1]
    accuracy <- (m - sum(incorrect.prediction))/m
    #print(accuracy)
    cv_10_fold(lda.fit,entire_aa)
}

#10 time 10 fold cross evaluation for errors using training set as CV data:


cat("Classification accuray of each class using 10-fold evaluation 10 times using entire dataset as testing set:\n")
for (i in 1:6) 
    {
    training_aa=raw_training[, !(colnames(raw_training) %in% subset_emotions[i,])] 
    testing_aa=raw_testing[, !(colnames(raw_testing) %in% subset_emotions[i,])]
    entire_aa=raw_entire[, !(colnames(raw_testing) %in% subset_emotions[i,])]
    cat("Emotion: ",class_names[i],"aka L",i,": ") 
    make_lda_model(training_aa,class_names[i],entire_aa,testing_aa)
}

cat("Classification accuray of each class using 10-fold evaluation 10 times using testing set as testing set:\n")
for (i in 1:6) 
    {
    training_aa=raw_training[, !(colnames(raw_training) %in% subset_emotions[i,])] 
    testing_aa=raw_testing[, !(colnames(raw_testing) %in% subset_emotions[i,])]
    entire_aa=raw_entire[, !(colnames(raw_testing) %in% subset_emotions[i,])]
    cat("Emotion: ",class_names[i],"aka L",i,": ") 
    make_lda_model(training_aa,class_names[i],testing_aa,testing_aa)
}
```
