The curve above is visible in Figures 4 and 5 in this paper,
where it is typical of the number of online news articles (and blog articles after this)

The paper tries to formalize the shape of the curve, but it doesn't finish the job -- it offers some equations, but no validation or justification.

The idea is compelling, though -- this kind of curve seems to describe the growth and decline of a population of instances of some Meme -- a phrase or sign that has meaning within a culture, and "spreads" by communication within the culture (http://en.wikipedia.org/wiki/Meme).

In other words, the life-cycle of baby names seems to look similar to the life-cycle of memes in news articles.
This makes sense if names are viewed as memes.


##Population Dynamics:  Competition in a World of Memes

The life-cycle curves in news articles track the <b>population size</b> of the articles.
At first the population shoots up, then levels off, and starts to drop steeply.
Other news memes fight for attention, and older memes lose their staying power.
Eventually all instances of the older meme can disappear.

You have heard of "Predator-Prey models'. 
The field of Population Dynamics (http://en.wikipedia.org/wiki/Population_dynamics) develops mathematical models of processes like this.  The curve above resembles that of a prey that initially multiplies rapidly but is then overcome by predators.
A popular mathematical formulation using differential equations is known as the Lotka-Volterra equation
(http://en.wikipedia.org/wiki/Lotka%E2%80%93Volterra_equation).
Predator-prey models measure the effect of resources (like food, or in new cycles, like attention of readers) on population sizes.
Limited resources cause competition, which can then lead to large increases and decreases in population size.

The point here is that the curve shape above might somehow be modeled by a curve that arises in population dynamics,
and there might actually be a justifiable mathematical model for these curves.
This is the point of departure for this problem:  is there a model for the life-cycle of baby names?

# Getting access to Google Trends or N-Grams Data

##Google Trends (and Correlate) -- time series summaries of Google web searches

Google Trends (http://www.google.com/trends/) permits analysis of time-series data about frequencies of specific search terms.
By definition, search terms are an excellent reflection of memes.  Resulting data is obtainable easily (see below).

Google Correlate (https://www.google.com/trends/correlate) permits retrieval of all time series that match (are highly correlated with) a given user-input time series.  Getting resulting data seems not as easy for some reason, and there seems no documented API.

Data from both of these start in 2004, and reach up to the present.

```python
# Adapted from:
# http://nbviewer.ipython.org/github/sahuguet/notebooks/blob/master/GoogleTrends%20meet%20Notebook.ipynb

import json
import re
import urllib2

%pylab inline

QUERY_TERMS = ['dollar', 'bitcoin']
QUERY = 'http://www.google.com/trends/fetchComponent?q=%s&cid=TIMESERIES_GRAPH_0&export=3' % ",".join(QUERY_TERMS)

data = urllib2.urlopen(QUERY).read()

# We need to do some data cleaning: remove Javascript header+footer,
# and translate new Date(....,..,..) into YYYY-MM-DD.
header = """// Data table response\ngoogle.visualization.Query.setResponse("""
data = data[len(header):-2]
data = re.sub('new Date\((\d+),(\d+),(\d+)\)', (lambda m: '"%s-%02d-%02d"' % (m.group(1).strip(), 1+int(m.group(2)), int(m.group(3)))), data)

data = data.replace(',,',',') ## bug in conversion

timeseries = json.loads(data)

initial_colnames = [col['label'] for col in timeseries['table']['cols']]
initial_rows = [[field['v'] for field in row['c']] for row in timeseries['table']['rows']]

colnames = ['year','month'] + initial_colnames[1:3]
rows = [([int(d) for d in row[0].split('-')[0:2]] + row[1:3]) for row in initial_rows][:-1]

print colnames
print rows

# to communicate these to R, we have to convert from the default dict structure.
```

```python
%%R -i rows,colnames -w 800 -h 400

# we need to use the '-i' form here to get all plots overlaid
nrows = nrow(rows)
times = rows[,1]+(rows[,2]-1)/12

plot( times, rows[,3], col="blue", type="l", ylim=c(0,100),
    main="We can use R for time-series analysis here", cex.main=2)
points( times, rows[,4], col="green4", type="l")
legend("topleft", colnames[3:4], col=c("blue","green4"), lwd=3 )
```

```python
import pandas as pd

#  you can do the same thing in Pandas/Matplotlib, instead of R

columns = [k['label'] for k in timeseries['table']['cols']]
simple_rows = map(lambda x: [k['v'] for k in x['c']], timeseries['table']['rows'])
dataframe = pd.DataFrame(simple_rows, columns=columns)
dataframe.set_index('Date', inplace=True)
dataframe.plot(figsize=(16, 8))
```
