```python
%%R

dpareto=function(x, shape=1, location=1) shape * location^shape / x^(shape + 1);
ppareto=function(x, shape=1, location=1) (x >= location) * (1 - (location / x)^shape);
qpareto=function(u, shape=1, location=1) location/(1 - u)^(1 / shape);
rpareto=function(n, shape=1, location=1) qpareto(runif(n), shape, location);
plot(function(x) {dpareto(x, shape = 1)}, 1, 3, ylab="power law density")
plot(function(x) {dpareto(x, shape = 1.5)}, 1, 3, col="red", add=TRUE)
plot(function(x) {dpareto(x, shape = 2)}, 1, 3, col="blue", add=TRUE)
plot(function(x) {dpareto(x, shape = .33)}, 1, 3, col="blue", add=TRUE)
legend(x = "topright", legend = c("alpha = 1", "alpha = 1.5", "alpha = 2"),  lty = c(1, 1, 1), col = c("black", "red", "blue"))
```

```python
%%R

# Function to determine location parameter of pareto dist using MLE estimate from data
location = function(d, max=100)  {
    stat = NULL;
    p = NULL;
    mstat = 1;
    mp = NULL;
    mlocation = NULL;
    mshape = NULL;
    for (location in 1:max) { 
        d = d[d >= location];
        # MLE for shape parameter (discrete case)
        shape = 1 / mean(log(d / (location - 0.5)));
        t = ks.test(d, "ppareto", shape = shape, location = location);
        stat[location] = t$stat;
        p[location] = t$p;
        if (stat[location] < mstat) {
             mstat = stat[location];
             mp = p[location];
             mlocation = location;
             mshape = shape;
        }
    }
    return(list(stat=mstat, p=mp, shape=mshape, location=mlocation, statvec=stat, pvec=p));
} 
```

```python
%%R
fit_comparisions1 = function(data) {
    meanlog = mean(log(data));
    sdlog = sd(log(data));
    par = location(data, 100);
    shape = par$shape;
    location = par$location;
    exp <-fitdistr(data,"exponential")
        pareto_test = ks.test(data[data>=location], "ppareto", shape = shape, location = location);
    lognormal_test = ks.test(data, "plnorm", meanlog = meanlog, sdlog = sdlog);
    exp_test = ks.test(data, "pexp", exp$estimate);
    
    cat("lognormal fit p-value: ");print(lognormal_test$p.value)
    cat("exp fit p-value: ");print (exp_test$p.value)
    cat("Pareto fit ksp p-value:");print(pareto_test$p.value);
      maxpvalue = max(lognormal_test$p.value, exp_test$p.value, pareto_test$p.value)
    if (maxpvalue == lognormal_test$p.value) 
        {
        cat("lognormal has the highest ks test p-value: ");print(lognormal_test$p.value)
        return(1)
        }
    else if (maxpvalue==exp_test$p.value) 
        {
        cat("exp has the highest ks test p-value: ");print(exp_test$p.value)
        return(2)
        }
    else 
        {
        cat("pareto has the highest ks test p-value: ");print(pareto_test$p.value)
        return(3)
        }
  

}
Jerry_count = c(25,59,149,392,781,1121,1679,2053,2654,2524,1709,1190,849,583,509,330,241,189,120,76,38,18,8)
a=fit_comparisions1(Jerry_count)
```

##Population Dynamics:  Competition in a World of Memes

The life-cycle curves in news articles track the <b>population size</b> of the articles.
At first the population shoots up, then levels off, and starts to drop steeply.
Other news memes fight for attention, and older memes lose their staying power.
Eventually all instances of the older meme can disappear.

You have heard of "Predator-Prey models'. 
The field of Population Dynamics (http://en.wikipedia.org/wiki/Population_dynamics) develops mathematical models of processes like this.  The curve above resembles that of a prey that initially multiplies rapidly but is then overcome by predators.
A popular mathematical formulation using differential equations is known as the Lotka-Volterra equation
(http://en.wikipedia.org/wiki/Lotka%E2%80%93Volterra_equation).
Predator-prey models measure the effect of resources (like food, or in new cycles, like attention of readers) on population sizes.
Limited resources cause competition, which can then lead to large increases and decreases in population size.

The point here is that the curve shape above might somehow be modeled by a curve that arises in population dynamics,
and there might actually be a justifiable mathematical model for these curves.
This is the point of departure for this problem:  is there a model for the life-cycle of baby names?

```python
import urllib2 
zipped_data = urllib2.urlopen("http://www.ssa.gov/oact/babynames/names.zip")
## this .zip file unpacks into a collection of files with names for each year 1880 - 2012.

import zipfile
from cStringIO import StringIO

zfiledata = StringIO(zipped_data.read())
zfile = zipfile.ZipFile(zfiledata)
for filename in zfile.namelist():
   if filename.endswith(".txt"):
      # We have a .txt file within the .zip file
      filedata = zfile.open(filename).read()
      print 'baby name file %s -- has length %d' % (filename, len(filedata))
      
```

```python
import re
import glob
import pandas as pd
import numpy as np

# Assemble all of the data into a single DataFrame
years = range(1880, 2011)
pieces = []
columns = ['name', 'sex', 'births']

yob_filename_re = re.compile(r"yob(\d{4})")

yobs = glob.glob("names/yob*.txt")

for path in yobs:
    year = yob_filename_re.findall(path)[0]
    frame = pd.read_csv(path, names=columns)
    frame['year'] = int(year)
    pieces.append(frame)

# Concatenate everything into a single DataFrame
names = pd.concat(pieces, ignore_index=True)
```

```python
def add_prop(group):
    births = group.births.astype(float)
    group['prop'] = births / births.sum()
    return group
names = names.groupby(['year', 'sex']).apply(add_prop)
names.head()
```

The idea is compelling, though -- this kind of curve seems to describe the growth and decline of a population of instances of some Meme -- a phrase or sign that has meaning within a culture, and "spreads" by communication within the culture (http://en.wikipedia.org/wiki/Meme).
